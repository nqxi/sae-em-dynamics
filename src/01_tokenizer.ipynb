```python
# https://huggingface.co/docs/transformers/en/main_classes/tokenizer
tokenizer = AutoTokenizer.from_pretrained("Qwen/Qwen2.5-1.5B-Instruct")
ids = tokenizer("Hello world")["input_ids"]
print(ids)
text = tokenizer.decode(ids)
print(text) # Hello world
```

```python
# https://huggingface.co/docs/transformers/v5.5.0/en/chat_templating#using-applychattemplate
messages = [
      {"role": "user", "content": "What is 2+2?"},
      {"role": "assistant", "content": "4."},
  ]
formatted = tokenizer.apply_chat_template(messages, tokenize=False)
print(formatted) 

# The special tokens let the model know when the user is speaking and when to respond
```
